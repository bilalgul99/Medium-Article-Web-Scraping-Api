```python
#https://colab.research.google.com/drive/1CEeS2rjK00jspuxPCpvZoEM3XQBe31PM?usp=sharing
#for API search got to http://bilalgul.pythonanywhere.com/
#to make API request for a json response http://bilalgul.pythonanywhere.com/search?keyword=<your_keywrord>
#code found at https://github.com/bilalgul99/Medium-Article-Web-Scraping-Apiimport urllib.request
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import json
data = pd.read_csv('url_technology.csv')

titles = []
subtitles = []
articles = []
authors = []
author_urls = []
claps = []
reading_times = []
image_sources = []
scrapped_urls=[]
for i in range(500):
    try:
        url_to_scrape = data.iloc[i]['url']
        r = requests.get(url_to_scrape, timeout=10)        
        text_got = r.text+''        
        r.close
        
        soup = BeautifulSoup(text_got, "html.parser")
        
        # Get the subtitle
        subtitle = soup.find('h2').text.strip()
        dataj = json.loads(soup.select_one('[type="application/ld+json"]').contents[0])
        #print(json.dumps(data, indent=4))
        print("Author:", dataj["author"]["name"])
        #print("URL:", dataj["author"]["url"])
        # Get the article
        article = ''
        for p in soup.find_all('p'):
            article += p.text.strip() + ' '
        
        
        if soup.find('button', class_='button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents'):
          clap = soup.find('button', class_='button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents').text
        else:
          clap = 0
        print(clap)
        
        # Get the claps
        #clap = soup.find('p', {'class': 'be b'}).contents[0]
        #print('second:'+clap)        

        #clap = soup.find('button', {'data-action': 'show-recommends'}).text.strip()
        
    
        # Get the claps
        #clap = soup.find('button', {'class': 'af ag ah ai aj ak al am an ao ap aq ar as at abg wh'}).contents[0]
        #clap=soup.find_all("button")[1].contents[0]
        #print(clap)
        #clap_count = soup.find('button', {'class': 'button button--chromeless u-baseColor--buttonNormal js-multirecommendCountButton u-disablePointerEvents'}).text.strip()
        claps.append(clap)
        #print(claps)
        # Get the reading time
        #print('time:'+soup.find('div', {'class':'ab ae'}).contents[0].strip())
        reading_time = soup.find('div', {'class':'ab ae'}).contents[0].strip()
        
        # Get the image source
        
        #adding data
        subtitles.append(subtitle)
        articles.append(article)
        authors.append(dataj["author"]["name"])
        author_urls.append(dataj["author"]["url"])
        titles.append(dataj["name"])
        image_sources.append(dataj["image"])
        reading_times.append(reading_time)
        scrapped_urls.append(url_to_scrape)
        print(f'i{i}')


    except Exception as e:
      print("An exception occurred:", e)
      print("no data found")

# Create a dataframe from the scraped data
data = {'Title': titles, 'Subtitle': subtitles,'Article Url':scrapped_urls, 'Article': articles, 'Author': authors, 'Author URL': author_urls, 'Claps': claps, 'Reading Time': reading_times, 'Image Source': image_sources}
df = pd.DataFrame(data)

# Save the dataframe to a CSV file
df.to_csv('medium_articles.csv', index=False)
```
